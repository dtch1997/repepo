```python
import subprocess
from repepo.experiments.caa_repro.utils.helpers import SteeringSettings

settings = SteeringSettings(
    type = "truthful_qa",
    model_size = "13b"
)

for layer in [16]:
    for multiplier in [-1.0, -0.5, 0, 0.5, 1.0]:
        save_suffix = settings.make_result_save_suffix(
            layer=layer, 
            multiplier=multiplier
        )
        save_path = f"results_{save_suffix}.json"
        # Replace '=' with unicode
        save_path = save_path.replace("=", "%3D")
        https_path = f"https://raw.githubusercontent.com/nrimsky/CAA/main/results/{save_path}"
        subprocess.run(f"wget {https_path}", shell=True)
```

```python
import json

def compute_average_correct_prob(result: list[dict[str, str]]):
    total_correct_prob = 0.0
    for r in result:
        if 'A' in r['correct']:
            total_correct_prob += float(r['a_prob'])
        elif 'B' in r['correct']:
            total_correct_prob += float(r['b_prob'])
        else:
            raise ValueError(f"Invalid correct answer: {r['correct']}")
    return total_correct_prob / len(result)

for layer in [16]:
    for multiplier in [-1.0, -0.5, 0, 0.5, 1.0]:
        save_suffix = settings.make_result_save_suffix(
            layer=layer, 
            multiplier=multiplier
        )
        save_path = f"results_{save_suffix}.json"

        with open(save_path, "r") as f:
            result = json.load(f)
        print(compute_average_correct_prob(result))

        break
    break 
        

```
